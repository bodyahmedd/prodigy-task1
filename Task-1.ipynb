```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Abdelrahman\Downloads\Telegram Desktop\population-by-age-Group.csv")
print(df.head())
```

```python
ages=['Age: 65+','Age: 25-64','Age: 15-24','Age: 5-14','Age: 0-4']
age_total=[df[age].sum() for age in ages]
x={"age":ages,"age_total":age_total}
dff=pd.DataFrame(x)
print(dff)
```

```python
plt.figure(figsize=(10,6))
plt.bar(dff["age"],dff["age_total"],color="skyblue",edgecolor="red")
plt.title("Age Distribution in Population")
plt.xlabel("Age")
plt.ylabel("Total_Populatin")
plt.show()
```
